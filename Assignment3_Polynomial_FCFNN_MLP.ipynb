```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
import numpy as np
import matplotlib.pyplot as plt
```

```python

def main():
    model1, test1X, test1Y = prepare_model(1)
    evaluate_and_plot(test1X, test1Y, model1.predict(test1X))

    model2, test2X, test2Y = prepare_model(2)
    evaluate_and_plot(test2X, test2Y, model2.predict(test2X))

    model3, test3X, test3Y = prepare_model(3)
    evaluate_and_plot(test3X, test3Y, model3.predict(test3X))



def prepare_model(polynomial_number):
    model = build_model()
    model.compile(loss='mse')

    x, y = data_process(polynomial_number)

    (trainX, trainY), (valX, valY), (testX, testY) = prepare_train_val_test(x, y)
    validation_data = (valX, valY)
    model.fit(trainX, trainY, validation_data=validation_data, epochs=30)

    return model, testX, testY


def evaluate_and_plot(testX, testY, y_pred):
    sorted_idx = np.argsort(testX.reshape(-1))
    testX_sorted = testX[sorted_idx]
    testY_sorted = testY[sorted_idx]
    y_pred_sorted = y_pred[sorted_idx]

    plt.figure(figsize=(8,6))
    plt.scatter(testX_sorted, testY_sorted, color='blue', label='True values', alpha=0.5)
    plt.plot(testX_sorted, y_pred_sorted, color='red', label='Predicted value', linewidth=2)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Polynomial Regression with Neural Network")
    plt.legend()
    plt.show()


def prepare_train_val_test(x, y):
    total_n = len(x)
    train_n = int(total_n*0.7)
    val_n = int(total_n*0.1)
    test_n = int(total_n*0.2)

    trainX = x[: train_n]
    trainY = y[: train_n]

    valX = x[train_n : train_n+val_n]
    valY = y[train_n : train_n+val_n]

    testX = x[train_n+val_n :]
    testY = y[train_n+val_n :]

    print(len(trainX), len(valX), len(testX))
    return (trainX, trainY), (valX, valY), (testX, testY)


def data_process(polynomial_number):
    n = 5000
    x = np.random.randint(-n, n, n) #(left_limit, right_limit, count)
    y = []
    for i in range(n):
        y.append(my_polynomial(polynomial_number, x[i]))

    x = np.array(x).reshape(-1,1)
    y = np.array(y).reshape(-1,1)

    x = x / np.max(np.abs(x))
    y = y / np.max(np.abs(y))


    return x, y

def my_polynomial(polynomial_number, x):
    if (polynomial_number == 1):
        y = 5*x + 10
    elif (polynomial_number == 2):
        y = 3*x*x + 5*x + 10
    elif (polynomial_number == 3):
        y = 4*x*x*x + 3*x*x + 5*x + 10
    else:
        print("Polynomial Undefined")

    return y

def build_model():
    inputs = Input((1, ))
    h1 = Dense(8, activation = 'relu', name = 'hidden_layer1')(inputs)
    h2 = Dense(4, activation = 'relu', name = 'hidden_layer2')(h1)
    h3 = Dense(8, activation = 'relu', name = 'hidden_layer3')(h2)
    outputs = Dense(1, activation = 'linear', name = 'output_layer')(h3)
    model = Model(inputs, outputs)
    model.summary(show_trainable = True)
    return model

if __name__ == '__main__':
    main()
```
